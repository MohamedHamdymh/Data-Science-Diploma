```python
import seaborn as sns
import pandas as pd

%matplotlib inline
sns.set(rc={'figure.figsize': [9, 9]}, font_scale=1.2)
```

```python
df = pd.read_csv('50ulke.csv')
```

```python
df.head()
```

```python
sns.countplot(x='Continent', data=df)
```

```python
sns.set(font_scale=1.5)
```

```python
sns.catplot(x="Continent", y="National Income", kind="bar",hue="Countries", data=df,height=10, aspect=1)
```

```python
sns.boxplot(x='Continent', y='National Income', data=df)
```

```python
sns.stripplot(x='Continent', y='National Income', data=df)
```

```python
sns.stripplot(x='Continent', y='Growth', data=df)
```

```python
sns.boxplot(x='Continent', y='Growth', data=df)
```

```python
sns.catplot(x="Continent", y="Growth", kind="bar", data=df,height=8, aspect=1.2)
```

```python
sns.catplot(x="Countries", y="National Income", kind="bar", data=df,height=8, aspect=1.2)
```

```python
sns.set(font_scale=2)
```

```python
sns.catplot(x="Countries", y="Growth", kind="bar", data=df,height=25, aspect=2)
```

```python
sns.jointplot(x='Growth', y='Per person', data=df, kind='scatter')
```
