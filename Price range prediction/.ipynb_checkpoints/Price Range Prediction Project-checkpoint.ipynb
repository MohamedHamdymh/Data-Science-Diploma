5) Models Building

* splitting data

```python
x = df.drop('price_range',axis=1)
y = df['price_range']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.2, random_state=5)
```

- selecting best k for KNN model

```python
knn_scores = []
for k in range(1, 50):
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    score = accuracy_score(y_test, y_pred)
    knn_scores.append(score)
```

```python
np.argmax(knn_scores) + 1
```

- best k is 11

```python
np.max(knn_scores)
```

```python
plt.plot(range(1, 50), knn_scores)
```

this figure shows different accuracy for k range (0 : 50)

- Train The Model with k=11

```python
knn_model = KNeighborsClassifier(n_neighbors=np.argmax(knn_scores) + 1)
```

- Comparing different Classification algorithms

- Evaluate the Models

```python
models = [LogisticRegression(),LinearSVC(),SVC(kernel='rbf'),KNeighborsClassifier(n_neighbors=11),RandomForestClassifier(),
        DecisionTreeClassifier(),GaussianNB()]
model_names=['Logistic Regression','Linear SVM','rbf SVM','KNearestNeighbors','Random Forest Classifier','Decision Tree','Naive Base GaussianNB']

accuracy=[]
d={}

for model in models:
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    accuracy.append(accuracy_score(y_pred,y_test))
    print(model,end='\n')
    print(end='\n')
    print('confusion_matrix')
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test,y_pred))
    print('---------------------------------------------------------------------------------------')
    
     
d={'Model':model_names,'Accuracy':accuracy}
d
```

```python
accuracy_df=pd.DataFrame(d)
accuracy_df.sort_values(by = 'Accuracy', ascending = False)
```

```python
sns.barplot(y='Model',x='Accuracy',data=accuracy_df.sort_values(by = 'Accuracy', ascending = False))
```

- Using K-fold cross Validation

```python
kfold = KFold(n_splits=10)
cv_meanlist = []
accuracy=[]
models = [LogisticRegression(),LinearSVC(),SVC(kernel='rbf'),KNeighborsClassifier(n_neighbors=11),RandomForestClassifier(),
        DecisionTreeClassifier(),GaussianNB()]
model_names=['LogisticRegression','LinearSVM','rbfSVM','KNearestNeighbors','RandomForestClassifier','DecisionTree','GaussianNB']


for model2 in models:
    
    cv_score = cross_val_score(model2,x_train, y_train, cv = kfold, scoring = "accuracy")
    cv_meanlist.append(cv_score.mean())
    accuracy.append(cv_score)


cv_df=pd.DataFrame(cv_meanlist, index=model_names)   
cv_df.columns=['CV Mean']    
cv_df
cv_df.sort_values(['CV Mean'], ascending=[0])
```
