```python
import seaborn as sns
import pandas as pd

%matplotlib inline
sns.set(rc={'figure.figsize': [9, 9]}, font_scale=1.2)
```

```python
df = pd.read_csv('50ulke.csv')
```

```python
df.head()
```

```python
sns.countplot(x='Continent', data=df)
```

* this graph shows number of countries of every continent in the data.

```python
sns.catplot(y="Countries", x="National Income", kind="bar", data=df,height=10, aspect=1.2)
```

* This figure illustrates the national income (US dollar) of the countries.

```python
sns.catplot(y="Countries", x="Growth", kind="bar", data=df,height=10, aspect=1.2)
```

* This figure illustrates the growth percentage of the countries.

```python
sns.catplot(y="Countries", x="Per person", kind="bar", data=df,height=10, aspect=1.2)
```

* This figure illustrates the income per person (US dollar) of the countries.

```python
sns.stripplot(x='Continent', y='National Income', data=df)
```

* this graph shows that the national income of the majority of countries doesn't exceed 2500 million dollars

```python
sns.boxplot(x='Continent', y='National Income', data=df)
```

* This figure illustrates the mean of every continent

```python
sns.stripplot(x='Continent', y='Growth', data=df)
```

* This figure illustrates distribution of growth percentage of the countries of every continent

```python
sns.boxplot(x='Continent', y='Growth', data=df)
```

* This figure illustrates the mean of growth percentage  of every continent and the range of growth percentage of counytries in every continent

```python
sns.boxplot(x='Continent', y='Per person', data=df)
```

* This figure illustrates the mean of income of per person of every continent and the range of income of per person of counytries in every continent

```python
sns.jointplot(y='National Income', x='Per person', data=df, kind='scatter')
```

* comparsion between national income and income of per person and their distribution
